```python
#Import request libray
import requests
```

```python
standings_url = "https://fbref.com/en/comps/9/Premier_League-Stats"
```

```python
#Using request to download the html of the url
data = requests.get(standings_url)
```

```python
#inspect the html
data.text
```

###### We need to Parse our HTML Scripts. To do this we import BeautifulSoop

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data.text)
```

```python
standings_table = soup.select('table.stats_table')[0]
```

```python
standings_table
```

```python
# To all the a tags in the "table.stats_table"
links = standings_table.find_all('a')
```

```python
'''To get the href property of each table, to to this we will a list 
comprehension '''
links = [l.get("href") for l in links]
```

```python
links = [l for l in links if '/squads/' in l]
```

```python
links
```

```python
''' The links only have the last part of the link, they don't have 
the domain added to it, so we need to add the domain to form a full url'''

team_urls = [f"https://fbref.com{l}" for l in links]
```

```python
team_urls
```

#### Extract the Match Stats Using Pandas and Requests

```python
'''We need to work with the team stats, so first we will 
consider Man_city'''
team_url = team_urls[0]
```

```python
data = requests.get(team_url)
data.text
```

```python
# we need the scores and fixtures table
import pandas as pd

matches = pd.read_html(data.text, match = "Scores & Fixtures")
```

```python
matches[0]
```

```python
shooting = pd.read_html(data.text, match = "Shooting")[0]
```

#### Cleaning and Merging Scrapped Data with Pandas

```python
shooting.head()
```

```python
shooting.columns = shooting.columns.droplevel()
```

```python
shooting.head()
```

```python
# We need to combine the shooting dataframe to the match match DataFrame (not all columns)
team_data = matches[0].merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
```
