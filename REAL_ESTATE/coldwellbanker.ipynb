```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json
from lxml import etree
```

```python
main_url = 'https://www.coldwellbanker.aw/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:47.0) Gecko/20100101 Firefox/47.0',
    "Accept-Encoding": "*",
    "Connection": "keep-alive"
    }
hyper = ['https://www.coldwellbanker.aw/index.php?action=searchresults&pclass[]=1&pclass[]=2&pclass[]=6&pclass[]=4',
         'https://www.coldwellbanker.aw/index.php?cur_page=1&action=searchresults&pclass[]=1&pclass[]=2&pclass[]=6&pclass[]=4&sortby=price&sorttype=DESC',
         'https://www.coldwellbanker.aw/index.php?cur_page=2&action=searchresults&pclass[]=1&pclass[]=2&pclass[]=6&pclass[]=4&sortby=price&sorttype=DESC'
    'https://www.coldwellbanker.aw/index.php?cur_page=3&action=searchresults&pclass[]=1&pclass[]=2&pclass[]=6&pclass[]=4&sortby=price&sorttype=DESC',
    'https://www.coldwellbanker.aw/index.php?cur_page=4&action=searchresults&pclass[]=1&pclass[]=2&pclass[]=6&pclass[]=4&sortby=price&sorttype=DESC',
    'https://www.coldwellbanker.aw/index.php?cur_page=5&action=searchresults&pclass[]=1&pclass[]=2&pclass[]=6&pclass[]=4&sortby=price&sorttype=DESC',
    'https://www.coldwellbanker.aw/index.php?action=searchresults&pclass[]=5&pclass[]=10&pclass[]=11#',
    'https://www.coldwellbanker.aw/index.php?cur_page=1&action=searchresults&pclass[]=5&pclass[]=10&pclass[]=11&sortby=price&sorttype=DESC'
    ]
```

```python
def get_all_url(li):
    all_links = []
    for url in li:
        req = requests.get(url)
        soup = BeautifulSoup(req.content, "html.parser")
        anchors = soup.find_all('div', {'class': "col-sm-9"})
    #print(anchors)
        for anchor in anchors:
            links  = anchor.find_all('a', {'class':'btn btn-primary btn-darkblue','href' :True})
            for link in links:
                urls = main_url + link.get('href')
                print(urls)
                all_links.append(urls)
    return all_links
path = get_all_url(hyper)
```

```python
def get_details(urls):
    property_list = []
    for link in urls:
        req = requests.get(link, headers = headers)
        soup = BeautifulSoup(req.content, 'html.parser')
        dom = etree.HTML(str(soup))
        name = soup.find('h1', class_ = "listing_title").text
        #rooms = dom.xpath('//*[@id="layout-content"]/div[3]/div/div/div[1]/div/div[2]/div/div/div[1]/ul/li[3]/div/div[2]')[0].text.strip()
        #bathroom = dom.xpath('//*[@id="layout-content"]/div[3]/div/div[2]/div[1]/div/div[2]/div/div/div[1]/ul/li[4]/div/div[2]')
        try:
            test1 = dom.xpath('/html/body/div[3]/div/div/div[2]/div/div[1]/div[1]/div/div[3]/div[7]/div/div[2]')[0].text.strip()
        except IndexError:
            test1 = dom.xpath('/html/body/div[3]/div/div/div[2]/div/div[1]/div[1]/div/div[3]/div[5]/div/div[2]')[0].text.strip()
        if (any(map(str.isdigit, test1))) == True:
            try:
                location  = dom.xpath('/html/body/div[3]/div/div/div[2]/div/div[1]/div[1]/div/div[3]/div[11]/div/div[2]')[0].text.strip()
            except:
                location = dom.xpath('/html/body/div[3]/div/div/div[2]/div/div[1]/div[1]/div/div[3]/div[8]/div/div[2]')[0].text.strip()
            rooms = dom.xpath('/html/body/div[3]/div/div/div[2]/div/div[1]/div[1]/div/div[3]/div[7]/div/div[2]')[0].text.strip()
            bathroom  = dom.xpath('/html/body/div[3]/div/div/div[2]/div/div[1]/div[1]/div/div[3]/div[8]/div/div[2]')[0].text.strip()
        else:
            location = test1
            rooms  = None
            bathroom = None
        #pool = dom.xpath('//*[@id="layout-content"]/div[3]/div/div[2]/div[1]/div/div[2]/div/div/div[2]/ul/li[5]/div/div[2]')
        property_status = dom.xpath('/html/body/div[3]/div/div/div[2]/div/div[1]/div[1]/div/div[3]/div[4]/div/div[2]')[0].text.strip()
        price =  soup.find('span', class_ = "listing_price").text
        #property_size = soup.find()
        

            
        prop = {
            'name': name,
            'location': location,
            'property': property_status,
            'price': price,
            'bedrooms': rooms,
            'bathrooms' : bathroom,
            'Pool':None,
            'link': link
        }
        property_list.append(prop)
        print('Saving:', prop['name'])
        
    df = pd.DataFrame(property_list)
    print(df.head(20))
    df.to_csv('coldwellbanker.csv', header=True, index=False)
```
